```python
import pandas as pd
def emi(amount ,year,interest_rate):
    interest_rate = interest_rate/100
    month = year * 12
    monthly_rate = interest_rate/12
    monthly_multiplier = 1+monthly_rate
    emi_amount = amount* monthly_rate *((monthly_multiplier**month)/(monthly_multiplier**month-1))
    return emi_amount
def amortization_table(amount,year,interest_rate):
    month = year*12
    data = pd.DataFrame(columns = ['month','emi','interest','principle','balance'])
    data['month'] = range(1,month)
    data['emi'] = emi(amount,year,interest_rate)
    for i in range(month):
        print(data.loc[i])
        
amortization_table(300000,20,5)
    

```
