```python
import pandas as pd
import plotly.express as pe
import plotly.graph_objects as go 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from scipy.stats import skew,kurtosis
final_data = pd.read_excel('final.xlsx')
columns = final_data.columns
df_list = np.array(final_data[columns[3:24]].values.tolist()).flatten()
std = np.std(df_list)
mean = np.mean(df_list)
min_value = mean -3*std
max_value = mean +3*std
outlier_data = final_data[(final_data[columns[3:24]] < min_value).any(axis=1) | (final_data[columns[3:24]] > max_value).any(axis=1)]
final_data = final_data[~(final_data[columns[3:24]] >= max_value).any(axis=1)]
final_data = final_data[~(final_data[columns[3:24]] <= min_value).any(axis=1)]
# final_data['Remarks'] = final_data['Remarks'].map({'Constant':'good','Goodnews':'good','Badnews':'bad'})
final_data.dropna(how = 'any',subset = ['Remarks','EPS'],inplace=True)
new_df_list = np.array(final_data[columns[3:24]].values.tolist()).flatten()
```

```python
#Below code is for good and bad news
for remarks  in final_data.Remarks.unique() :
    print(remarks)
    remark_data = final_data.query('Remarks == @remarks')
    df_list =  np.array(remark_data[columns[3:24]].values.tolist()).flatten()
    pe.histogram(df_list).show()
    skewness = skew(df_list)
    kurt = kurtosis(df_list)
    print(f'skewness: {skewness}, kurtosis: {kurt}')
```

```python
# THis code is for DM1 DM2 BM1  and BM2
for remarks  in final_data.market_condition.unique() :
    print(remarks)
    remark_data = final_data.query('market_condition == @remarks')
    df_list =  np.array(remark_data[columns[3:24]].values.tolist()).flatten()
    pe.histogram(df_list).show()
    skewness = skew(df_list)
    kurt = kurtosis(df_list)
    print(f'skewness: {skewness}, kurtosis: {kurt}')
```
