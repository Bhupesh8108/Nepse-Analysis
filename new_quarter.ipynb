```python
import pandas as pd
import os
old_companies = os.listdir('my_data')
# new_companies = os.listdir('')
# for old_comp in companies:
old_comp = 'NEPSE.xlsx'
old_comp_data = pd.read_excel(f'my_data/{old_comp}',skiprows=1)
new_comp_data = pd.read_excel(f'add_on_data/{old_comp}',skiprows=1)
merged_data = pd.concat([new_comp_data,old_comp_data],join= 'inner',axis = 0)
merged_data.drop_duplicates(subset=['Date'],inplace=True)
merged_data.sort_values(by = 'Date',ascending=True,inplace = True)
merged_data.to_excel(f'merged_data/{old_comp}',index = False)


    
```
